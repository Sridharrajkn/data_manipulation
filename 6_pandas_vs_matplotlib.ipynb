```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("P87-S2-Dataset-Basics-Resources/heart.csv")
df.head(2)
```

```python
fig,axes = plt.subplots(ncols=2)
df.plot.scatter(x="age", y="chol",ax=axes[0],color="red")
df.plot.scatter(x="age", y="trestbps",ax=axes[1])
fig.tight_layout();
```

```python
fig2 = axes[0].get_figure()
fig2.savefig("plots.png",bbox_inches="tight", transparent=True)
```

```python
with plt.style.context("dark_background"): #with plt.style.context("seaborn")
    fig, axes = plt.subplots(ncols=3, sharey=True, gridspec_kw={"width_ratios":[2, 1, 1], "wspace":0})
    y="age"
    xs=["chol", "trestbps", "thalach"]
    for x, ax in zip(xs,axes):
        ax.scatter(df[x], df[y])
        ax.set_xlabel(x)
    axes[0].set_ylabel(y)
    fig.savefig("output.png", dpi=100, bbox_inches="tight")#change dpi=(1000) for more clarity ; check the size of the file
    fig.savefig("output.pdf", dpi=600, bbox_inches="tight")
```
